# Seaborn




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#sample data
data = {
    "Students":["Aarav","Divya","Riyas","Anu","Arjun"],
    "Fees Paid":[18000,43200,56000,38400,50000]
}
df = pd.DataFrame(data)

#Simple bar chart
sns.barplot(x="Students",y="Fees Paid",data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

fees=[48000,43200,45600,38000,48200]

density = gaussian_kde(fees)
xs = np.linspace(min(fees),max(fees),200)
plt.plot(xs,density(xs))
plt.show()
```

```python
Fees:[18000,43200,56000,38400,50000]
#Simple box plot
plt.boxplot(fees)
plt.show()
```

```python
Fees:[18000,43200,56000,38400,50000]
#Simple box plot
plt.hist(fees)
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["Gopi","Lazar","Srinivasan","Jayabalan","Vajjiravelu"]
fees = [48000, 43200, 45600, 38400, 48000]
plt.barh(students, fees, color="lightgreen")
plt.title("Fee Paid by students")
plt.xlabel("fees")
plt.ylabel("students")
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["Gopi","Lazar","Srinivasan","Jayabalan","Vajjiravelu"]
fees = [48000, 43200, 45600, 38400, 48000]
colors = ["lightblue", "lightcoral", "lightgreen", "gold", "plum"]
plt.pie(fees, labels=students, colors=colors)
plt.title("Fee Paid by students")
plt.show()
```

```python
import plotly.express as px
import pandas as pd

#Sample data
data={
    "Students":["Aarav","Divya","Riyas","Anu","Arjun"],
    "Fees Paid":[18000,43200,56000,38400,50000]
}
df=pd.DataFrame(data)

#Create bar chart
fig=px.bar(df,x="Students",y="Fees Paid",title="Fee Paid by Students")

#Save to HTML file (opens directly in browser, no server)
fig.write_html("fee_chart.html", auto_open=True)
```
